# Pasos para instalar Gitlab en Ubuntu 20.04

#### La primera forma de instalacion es haciendo una actualizacion de paquetes dentro del mismo ubuntu utilizando la consola y haciendo las siguientes actualizaciones

- Primero hay que hacer un Update dentro de la consola y confirmar los pasos de administrador agregando su contraseña para confirmar las actualizaciones

```python
import subprocess
from os import system
system("sudo apt-get update")
```

- Despues de que haya terminado de hacer el update hay que proceder hacer el siguiente paso que es Upgradear todos los paquetes

```python
from os import system
system("sudo apt-get upgrade")
```

- Despues de hacer lo anterior procederemos hacer las instalaciones de los certificados y el ssh 

```python
from os import system
system("sudo apt install ca-certificates curl openssh-server postfix")
```

- Ingresamos a la carpeta raiz de tpm y despues ahi ejecutamos el comando para bajar el repositorio

```python
from subprocess import check_output
comando = "cd /tmp && curl -LO https://packages.gitlab.com/install/repositories/gitlab/gitlab-ce/script.deb.sh"
```

- Ejecutaremos el siguiente Script despues de bajar el repositorio para ingresar a la raiz de TMP

```python
from os import system
system("sudo bash /tmp/script.deb.sh")
```

- Despues de que se baje el repositorio procederemos a instalar Gitlab

```python
from os import system
system("sudo apt install gitlab-ce")
```

#### Si se hizo correctamente los pasos anteriores deberia de aparecer en la consola el logo de gitlab con caracteres, si no, verificar unos de los procesos anteriores que se hiciera correctamente

- si todo esta correcto procederemos a configurar el firewall para dar de alta HTTP, HTTPS y SSH

```python
from subprocess import check_output
comando = "sudo ufw allow http && sudo ufw allow https && sudo ufw allow ssh"
```

- Ya para finalizar haremos una reconfiguracion del repositorio para dar de alta y ver el sistema de gitlab en accion

```python
from os import system
system("sudo gitlab-ctl reconfigure")
```

#### Si todo se hizo correctamente podran ver gitlab ya habilitado en una ventana de firefox utilizando en el URL: 'localhost/'

#### EL siguiente paso sera ingresar a nuestro repositorio de gitlab y para eso necesitamos un usuario y contraseña y para eso necesitamos saberlo, bien lo principal de saber que en la reconfiguracion que hicimos del gitlab por defecto el usuario es "root" y la contraseña esta cifrada, me preguntaran que de que sirve tener gitlab si no tenemos la contraseña, bien para eso hay que saber que en automatico gitlab la crea y la deposita en una carpeta oculta, asi que lo siguiente que haremos es abrir la terminal y buscarla

- Bien para poder a llevar a cabo el usuario y contraseña sabemos que el usuario por defecto es root, pero ahora lo siguiente que haremos es ver donde esta la contraseña y para ello necesitaremos irnos a la consola y buscarla; esta contraseña por defecto crea una carpeta llamada gitlab.
asi que lo primero a hacer es ir a la carpeta antiguamente mencionada donde si pusieron atencion a la instalacion nos da una ruta donde se instalo que en este caso es '/etc/gitlab' donde aqui es donde esta los archivos de instalacion y donde se situa la contraseña asi que abriremos una terminal y nos dirijimos a esa carpeta

```python
from os import system
system("cd /etc/gitlab")
```

- Ya situados en la carpeta si ejecutamos el comando 'ls' veremos diferentes archivos de la cual una es de nuestra importancia, llamado "initial_root_password" pero si intentamos abrirlo como texto se daran cuenta que esta cifrado y no podremos hacer nada, asi que, lo siguiente hacer es quitar ese cifrado en automatico, por lo cual debemos poner el siguiente comando

```python
from os import system
system("chmod ugo+rwx initial_root _password")
```

- Si hicimos todo correcto ahora podremos buscarlo desde el explorador buscando la carpeta /etc despues buscar /gitlab y abrir el initial o si queremos abrirlo desde la terminal ejecutaremos el siguiente comando: 

```python
from os import system
system("sudo nano initial_root_password")
```

- A continuacion nos pedira la contraseña nuevamente de nuestro usuario y despues de hacerlo podran ver su contraseña, la copiamos y la pegamos en nuestra pagina de gitlab, ya una vez hecho tendremos acceso a gitlab y a disfrutar!
